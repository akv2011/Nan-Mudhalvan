## Importing **Libraries**

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread
import seaborn as sns
```

## Assigning  path for **Train** and **Test** images

```python
train_path = dir_path +'/Train'
test_path = dir_path + '/Test'
```

```python
print(sorted(os.listdir(train_path)))
```

```python
sorted(os.listdir(test_path))
```

## Visualizing 25 random sample images from test set

```python
import random

images_path = os.listdir(test_path)

plt.figure(figsize=(25,25))


for i in range(1,26):

    plt.subplot(5,5,i)
    random_img_path = test_path +'/'+ random.choice(images_path)
    rand_img = imread(random_img_path)
    plt.imshow(rand_img)
    plt.xlabel(rand_img.shape[1], fontsize = 20)
    plt.ylabel(rand_img.shape[0], fontsize = 20)
```

## Mean of the dimensions of all the images in training set.

```python
dim1 = []
dim2 = []

for i in range(0,43):
    labels = train_path + '/{0}'.format(i)
    image_path = os.listdir(labels)
    for x in image_path:
        img = imread(labels + '/' + x)
        dim1.append(img.shape[0])
        dim2.append(img.shape[1])
```

```python
data = pd.DataFrame({'dim1': dim1, 'dim2': dim2})
sns.jointplot(x='dim1', y='dim2', data=data)
plt.show()
```



## Mean Dimension 1

```python
np.mean(dim1)
```

## Mean Dimension 2

```python
np.mean(dim2)
```
